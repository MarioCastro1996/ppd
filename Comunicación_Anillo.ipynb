```python
%%writefile comunicacion_anillo.c
#include "mpi.h"
#include <stdio.h>
#include <string.h>

void main () {
    MPI_Init(NULL, NULL);

    int rank, size, fuente, destino;
    char mensaje[50];
    
    MPI_Comm_rank(MPI_COMM_WORLD,&rank);
    MPI_Comm_size(MPI_COMM_WORLD,&size);

    fuente=rank-1;
    destino=rank+1;

    if(rank==0){
        fuente=size-1;
    }
    if(rank==size-1){
        destino=0;
    }

    sprintf(mensaje, "Hola proceso %d un saludo desde el %d",destino,rank);
   
    MPI_Send(mensaje, strlen(mensaje)+1, MPI_CHAR, destino, 9, MPI_COMM_WORLD);
    MPI_Recv(mensaje, 50, MPI_CHAR, fuente, 9, MPI_COMM_WORLD,MPI_STATUS_IGNORE);
    printf("Soy %d recibi <%s>\n", rank, mensaje);
    
    MPI_Finalize();
    return ;
}
```

```python
%%script bash
mpicc comunicacion_anillo.c -o comunicacion_anillo
mpirun --allow-run-as-root -np 6 comunicacion_anillo
```
